### Central Limit Theorem Under Sampling Bias

```python
# Importing all neccessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

#### The Experiment:
* We will simulate rolling an unfair die where different outcomes (1-6) had unequal probabilities.
* For example, higher chances of rolling a 1, lower chances of rolling a 4 and following the same logic.
* We will create multiple random samples as a result of event : Rolling the biased die N times.

```python
event = np.array([1,2,3,4,5,6])
probs = np.array([32,12,18,9,23,6])
outcomes = np.array(np.repeat(event, probs))
```

#### The Distribution of the Data
* We will plot a PMF (probability mass function) of our event outcomes using histogram.
* Due to sampling bias, some numbers are systematically more likely to be selected in a sample than others, and we can notice the probability of being selected, here.

```python
outcomes
```

Notice that face 1 repeats so many times that it increases the chances of selecting 1 from these outcomes, and this frequency distribution can be concluded using following PMF.

```python
plt.figure(figsize=(5, 3))
plt.hist(outcomes, bins=np.arange(0.5, 7.5, 1), density=True, edgecolor='black', rwidth=0.8)
plt.title("PMF of Discrete Distribution (Unfair Die)")
plt.xlabel("Die Face")
plt.ylabel("Probability")
plt.xticks(event)
plt.tight_layout()
plt.show()
```

#### Simulation of the CLT
 * We will define number of samples and sample units for each sample.
 * All the sample units are selected from the distribution following above probability mass function.
 * We will draw random samples and plot their means to check if CLT holds or not.
 * **(By gradually increasing the sample size/number of samples)**

```python
sample_size = 30
num_samples = 100

samples = np.random.choice(outcomes, (num_samples, sample_size), replace = True)
means = np.mean(samples, axis = 1)

plt.hist(means, bins = 16)
plt.xlabel("Sample Means")
plt.ylabel("Density")
plt.show()
```

### Conclusion:
Despite the strong bias in individual die rolls, our experiment confirms that as sample size or number of samples increases, the distribution of sample means becomes approximately normal, demonstrating the Central Limit Theorem in action.
